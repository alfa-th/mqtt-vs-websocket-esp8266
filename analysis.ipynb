```python
import numpy as np
import pandas as pd
import json
```

```python
ws_active_5 = np.array(json.loads(open("node/data/ws-server-active-5.json", 'r').read())[1:5000])
ws_active_25 = np.array(json.loads(open("node/data/ws-server-active-25.json", 'r').read())[1:5000])
ws_active_50 = np.array(json.loads(open("node/data/ws-server-active-50.json", 'r').read())[1:5000])
ws_active_100 = np.array(json.loads(open("node/data/ws-server-active-100.json", 'r').read())[1:5000])
ws_active_200 = np.array(json.loads(open("node/data/ws-server-active-200.json", 'r').read())[1:5000])

data = {
    "5": ws_active_5,
    "25": ws_active_25,
    "50": ws_active_50,
    "100": ws_active_100,
    "200": ws_active_200
}

df = pd.DataFrame(data=data)
df
```

```python
boxplot = df.boxplot(showfliers=False)
boxplot.set_ylabel("Waktu (ms)")
boxplot.set_xlabel("Payload (bytes)")
boxplot.set_title("WS Active")
```
